import numpy as np

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
import matplotlib.pyplot as plt

def plot_data(df, title, ax):
    groups = df.groupby("group")
    for name, group in groups:
        ax.scatter(group.x1, group.x2, label=name, alpha=0.6)
    ax.set_title(title)
    ax.legend()

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

datasets = [df0_125, df0_25, df0_5, df1_0, df2_0]
titles = ["Noise: 0.125", "Noise: 0.25", "Noise: 0.5", "Noise: 1.0", "Noise: 2.0"]

for i, (df, title) in enumerate(zip(datasets, titles)):
    plot_data(df, title, axes[i])

plt.tight_layout()
plt.show()
````

As the noise level increases from 0.125 to 2.0, the clusters become progressively less distinct. At low noise (0.125, 0.25), the clusters are well-separated, making them easy to identify. However, as noise increases to 1.0 and 2.0, the clusters start to overlap significantly, making it difficult to distinguish clear boundaries.

````python
from sklearn.cluster import KMeans

def plot_kmeans(df, title, ax):
    kmeans = KMeans(n_clusters=3, random_state=100, n_init=10)
    df["cluster"] = kmeans.fit_predict(df[["x1", "x2"]])
    ax.scatter(df.x1, df.x2, c=df["cluster"], cmap="viridis", alpha=0.6)
    ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=100, label="Centers")
    ax.set_title(title)
    ax.legend()

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, (df, title) in enumerate(zip(datasets, titles)):
    plot_kmeans(df, f"KMeans {title}", axes[i])

plt.tight_layout()
plt.show()
````

In an ideal scree plot, the "elbow" point marks the optimal number of clusters, where adding more clusters provides diminishing returns in reducing inertia. At low noise, the scree plot likely shows a clear elbow at
k=3, indicating that three clusters are optimal. However, as noise increases, the data points become more spread out and overlapping, making the elbow less pronounced and causing uncertainty in selecting the best k.

````python
from sklearn.metrics import silhouette_score

def compute_silhouette(df):
    kmeans = KMeans(n_clusters=3, random_state=100, n_init=10)
    labels = kmeans.fit_predict(df[["x1", "x2"]])
    return silhouette_score(df[["x1", "x2"]], labels)

silhouette_scores = {title: compute_silhouette(df) for title, df in zip(titles, datasets)}
silhouette_df = pd.DataFrame.from_dict(silhouette_scores, orient="index", columns=["Silhouette Score"])

print(silhouette_df)
````

The elbow method works by finding a balance between reducing clustering distortion (within-cluster variance) and avoiding unnecessary complexity. In this simulation, at low noise, we see a sharp drop in inertia at
k=3, reinforcing that three clusters are optimal. However, as noise increases, the distinction between clusters blurs, leading to a more gradual decline in the scree plot, making it harder to pinpoint a definitive elbow. This highlights how noisy data can obscure the natural structure of clusters.
